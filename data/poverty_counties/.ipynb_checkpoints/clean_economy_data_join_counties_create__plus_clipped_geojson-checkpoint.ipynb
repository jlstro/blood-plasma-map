```python
import time
import os
import pandas as pd
from pandas.io.json import json_normalize
import json
pd.set_option('display.max_columns', None, 'display.max_colwidth', 500)
import numpy as np
```

```python
df_econ_raw = pd.read_csv('ACS_16_5YR_DP03.csv', encoding = 'latin1', low_memory = False)
```

```python
#check if import worked - lot's of stuff
df_econ_raw.head()
```

```python
#The explanation of what each column represents is in row 0
#make a new df from it, transpose and save. also drop row 0 from the raw data
#I opened this cheat sheet in Excel to have it side by side
df_explainer = df_econ_raw[0:1].T
df_explainer.to_csv('column_explainer.csv', index = True)
df_econ_raw = df_econ_raw[1:]
```

```python
#This is A LOT of data that the census site gave me. I need to get rid of most of them.
df_econ_raw.shape
```

```python
#From my Excel cheat sheet I know that the following might be of interest:
#index 9 -'HC03_VC04' - Percent; EMPLOYMENT STATUS - Population 16 years and over - In labor force
#index 131 - 'HC01_VC85' - Estimate; INCOME AND BENEFITS (IN 2016 INFLATION-ADJUSTED DOLLARS) - Total households - Median household income (dollars)
#index 269 -'HC03_VC131' - Percent; HEALTH INSURANCE COVERAGE - Civilian noninstitutionalized population - With health insurance coverage
#index 397 - 'HC03_VC171' -Percent; PERCENTAGE OF FAMILIES AND PEOPLE WHOSE INCOME IN THE PAST 12 MONTHS IS BELOW THE POVERTY LEVEL - All people
#I also need 0-2 to identify the counties. Drop all the others
df_econ = df_econ_raw[['GEO.id', 'GEO.id2', 'GEO.display-label', 'HC03_VC04', 'HC01_VC85', 'HC03_VC131', 'HC03_VC171']]
#rename the columns to make life easier
df_econ = df_econ.rename(columns={'GEO.id' : 'properties.county_code' , 'GEO.id2' : 'properties.county_code_short', 'GEO.display-label' : 'properties.name', 'HC03_VC04' : 'properties.employment_perc', 'HC01_VC85' : 'properties.houselhold_income_med', 'HC03_VC131' : 'properties.health_insurance_perc', 'HC03_VC171' : 'properties.poverty_level_perc'})
#split state into an extra cell for easier grouping
df_econ['properties.state'] = df_econ['properties.name'].str.extract(r",(.*)$")
df_econ['properties.name'] = df_econ['properties.name'].str.extract(r"(.*),.*$")
```

```python
#save it all into a csv
df_econ.to_csv('economy_by_counties_clean.csv', index = False)
```

```python
#now for the county geojson file I downloaded
with open('us_counties_raw.geojson') as json_data:
    geometry_data = json.load(json_data)
df_counties = pd.DataFrame.from_dict(json_normalize(geometry_data['features']), orient='columns')
```

```python
#check import
df_counties.head()
```

```python
#drop and rename columns
df_counties = df_counties.drop(df_counties.columns[[3, 4, 6, -1, -2]], axis=1)
df_counties = df_counties.rename(columns={'properties.AFFGEOID' : 'properties.county_code', 'properties.GEOID' : 'properties.county_code_short', 'properties.NAME' : 'properties.name_short' })
```

```python
df_counties.shape
```

```python
df_counties_eco = df_counties.merge(df_econ, on='properties.county_code')
```

```python
#some counties are lost :-(
df_counties_eco.shape
```

```python
#Realized that I forgot to add population in the original data set so add it here
df_pop = pd.read_csv('countypop.csv', encoding = 'latin1')
df_pop.head()
```

```python
#clean up, merge
df_pop = df_pop.drop(df_pop.index[0])
del df_pop['GEO.id2']
del df_pop['GEO.display-label']
df_counties_eco = pd.merge(df_counties_eco, df_pop, left_on = 'properties.county_code', right_on = 'GEO.id', how = 'left')
```

```python
#drop double column and rename (should've done this earlier...)
df_counties_eco = df_counties_eco.drop(df_counties_eco[['properties.county_code_short_y']], axis = 1)
df_counties_eco = df_counties_eco.drop(df_counties_eco[['GEO.id']], axis = 1)
df_counties_eco = df_counties_eco.rename(columns={'properties.county_code_short_x' : 'properties.county_code_short', 'respop72017' : 'properties.population'})
```

```python
#Add what I created in QGIS: count centers per county, calculate perc of county that is in 50-mile-radius of center
df_buffer = pd.read_csv('../qgis_work/counties_centers_clipped_joined_clean.csv')
```

```python
#merge into new df with all info in it
df_counties_eco_buffer = pd.merge(df_counties_eco, df_buffer, left_on = 'properties.county_code', right_on = 'AFFGEOID', how = 'left')
```

```python
#rename some, drop some, add some
df_counties_eco_buffer = df_counties_eco_buffer.rename(columns={'Zugeschnitten_area_full' : 'properties.area' , 'Zugeschnitten_area_clipped' : 'properties.area_clipped' , 'Anzahl_centers_in_counties' : 'properties.centers'})
del df_counties_eco_buffer['Unnamed: 0']
del df_counties_eco_buffer['AFFGEOID']
df_counties_eco_buffer['properties.area_clipped'] = df_counties_eco_buffer['properties.area_clipped'].str.replace('.', '')
df_counties_eco_buffer['properties.area_perc'] = df_counties_eco_buffer['properties.area_clipped'].astype(float) / df_counties_eco_buffer['properties.area']
df_counties_eco_buffer['properties.population_perc'] = df_counties_eco_buffer['properties.population'].astype(float)*df_counties_eco_buffer['properties.area_perc']
```

```python
#add some color!
def add_color(row):
    temp = {}
    
    employ = float(row['properties.employment_perc'])
    if employ < 45.5:
        temp['properties.employment_perc_color'] = '#013646'
    elif employ < 55.1:
        temp['properties.employment_perc_color'] = '#015764'
    elif employ < 63:
        temp['properties.employment_perc_color'] = '#017783'
    else:
        temp['properties.employment_perc_color'] = '#0198A1'
    
    health = float(row['properties.health_insurance_perc'])        
    if health < 79:
        temp['properties.health_insurance_perc_color'] = '#004318'
    elif health < 85:
        temp['properties.health_insurance_perc_color'] = '#016B29'
    elif health < 90:
        temp['properties.health_insurance_perc_color'] = '#6AAF21'
    else:
        temp['properties.health_insurance_perc_color'] = '#BCCD02'
    
    pov = float(row['properties.poverty_level_perc']) 
    if pov < 13:
        temp['properties.poverty_level_perc_color'] = '#700D10'
    elif pov < 20:
        temp['properties.poverty_level_perc_color'] = '#940D13'
    elif pov < 28:
        temp['properties.poverty_level_perc_color'] = '#EF7F01'
    else:
        temp['properties.poverty_level_perc_color'] = '#FDCC03'
    
    income = float(row['properties.houselhold_income_med'])    
    if income < 40784:
        temp['properties.houselhold_income_med_color'] = '#002458'
    elif income < 51980:
        temp['properties.houselhold_income_med_color'] = '#01388B'
    elif income < 68929:
        temp['properties.houselhold_income_med_color'] = '#0474BC'
    else:
        temp['properties.houselhold_income_med_color'] = '#009DE0'
    
    return pd.Series(temp)
```

```python
#join, check integrity
df_counties_eco_buffer = df_counties_eco_buffer.apply(add_color, axis = 1).join(df_counties_eco_buffer)
df_counties_eco_buffer
```

```python
#save to file
df_counties_eco_buffer.to_csv('counties_center_merged.csv', index = False)
```

```python
# most of this is stolen code
def to_geojson(dataframe):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in dataframe:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```

```python
#sort table the right way
counties_json = json.loads(df_counties_eco_buffer.to_json(orient='records'))
counties_geo = to_geojson(counties_json)
```

```python
#write to file
with open('geo-data_counties_center.geojson', 'w') as outfile:
    json.dump(counties_geo, outfile)
```
